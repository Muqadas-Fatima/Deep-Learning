```python
# non vectorized form
# features (x valyes)
features = np.random.randint(low=1, high=10, size=(8, 4))
# y target all rows but only third column means last column
y=features[:,3]

# features
X = features[:, :3]  # dim  (8,3)


weights= np.random.rand(3)  #dim (3,1)
biases=np.zeros(1)


y_pred=[]
for i in range(0,8):
  y_pred.append(np.dot(weights.T,X[i]))


# loss
loss=y-y_pred
y_pred





```

```python
#vectorized form it is faster ,no loop needed

y_p=X@weights+biases
y_p
```
